```python
# first, to standarize the columns i'll rename them using lower and replacing the whitespaces for underlines

# saving the old columns name in a variable
old_columns = attacks.columns

# using lower and replacing the whitespaces and anything that is not a word character or number for underlines
pattern = '[^a-zA-Z0-9]'
new_columns = [re.sub(pattern, "_", column.lower()) for column in attacks.columns]
print(new_columns)

# saving new columns in the dataframe and checking
attacks.columns = new_columns
```

```python
# fixing cases out of the desired, manually
cols = list(attacks.columns)

cols[23] = 'unnamed_23'
cols[22] = 'unnamed_22'
cols[9] = 'sex'
cols[12] = 'fatal_y_n'
cols[14] = 'species'
print(cols)

attacks.columns = cols
```

```python
# since i believe there are many empty rows, i'll drop the duplicates first
# saving the dups in a variable
attacks_dups = attacks.duplicated()

# creating a new variable for the clean df
attacks_drop = attacks.drop_duplicates()

# also eliminating columns unnamed_22 and unnamed_23, since they are not relevant
attacks_clean = attacks_drop.dropna(axis=1,thresh=10)

# now the dataframe is a lot better, an average of 6k non null on 6,3k entries
attacks_clean.info()
```

```python
# i'll drop the columns that i already know i'm not gonna be using in the analysis

cols2drop = ['investigator_or_source','pdf','href_formula','href','case_number_1','case_number_2','original_order','type','name','sex','age','injury']
df = attacks_clean.drop(columns=cols2drop)
df.head(2)
```

```python
# now i want to clean the most relevant columns for me: year, country, activity, fatal_y_n and species
```

```python
# treating most columns with strip and lower

df['country'] = df['country'].str.strip().str.lower()
df['activity'] = df['activity'].str.strip().str.lower()
df['fatal_y_n'] = df['fatal_y_n'].str.strip().str.lower()
df['species'] = df['species'].str.strip().str.lower()
df['area'] = df['area'].str.strip().str.lower()
df['date'] = df['date'].str.strip().str.lower()
df['case_number'] = df['case_number'].str.strip().str.lower()
```

```python
# fatal_y_n: row '2017' investigate
df[df['fatal_y_n'] == '2017']

# sub for n, since doesnt seem fatal. Where it is fatal, it is described in injury field
df['fatal_y_n'] = df['fatal_y_n'].replace(['2017'], 'n')

# i'll consider as mispelling of n, since its not described as fatal
df[df['fatal_y_n'] == 'm']
df['fatal_y_n'] = df['fatal_y_n'].replace(['m'], 'n')
df['fatal_y_n'].value_counts()
```

```python
# noticed a lot of input had the activity written in the text, so it can be deduce more or less

mask = df['activity'].str.contains('swimming') == True
df[mask]['activity'].value_counts()
```

```python
# treating those cases for the main activities
# important to notice that most variants will be unionized, like 'body surfing' will be only 'surfing'

df.loc[attacks_clean['activity'].str.contains('swimming') == True, 'activity'] = 'swimming'
df.loc[attacks_clean['activity'].str.contains('surfing') == True, 'activity'] = 'surfing'
df.loc[attacks_clean['activity'].str.contains('fishing') == True, 'activity'] = 'fishing'
df.loc[attacks_clean['activity'].str.contains('bathing') == True, 'activity'] = 'bathing'
df.loc[attacks_clean['activity'].str.contains('diving') == True, 'activity'] = 'diving'
```

```python
# noticed the same thing with species

df['species'].value_counts()
```

```python
# since there is a lot more optins for shark, i'll create a list and iterate over it

list_sharks = ['white shark','tiger shark','bull shark','blacktip shark','blue shark','bronze whaler','caribbean reef shark','grey nurse shark','grey reef shark','hammerhead shark','lemon shark','mako shark','oceanic whitetip shark','raggedtooth shark','sandtiger shark','sevengill shark','spinner shark','wobbegong shark','zambesi shark']

def species_cleaning(shark):
    try:
        for i in list_sharks:
            if i in shark:
                return i
            else:
                return shark
    except:
        return np.nan
    
    
df['species'].map(species_cleaning)
df['species'].value_counts()
```

```python
# with that cleaning done, i can analyze deeper
# count of incidents where activity is surfing, per country
# i wanted to double check that even only in surfing, USA was the country with most incidents

df[df['activity'] == 'surfing'].groupby(['country'])[['activity']].count().sort_values(by='activity',ascending=False)
```

```python
# i have created df with fishing and surfing season per area
# i want to import those here and merge to my df
# the season is 'fixed' so it'll be the same for each month, but it'll help me toa analyse if the count of incident are corelated to fishing or surfing

# creating a function to import google sheets by link

def read_from_gsheets(spreadsheet):
    """
    Transform url into csv 
    """
    working_spreadsheet = spreadsheet.replace('/edit?usp=sharing','/export?format=csv')
    
    return pd.read_csv(working_spreadsheet)

# testing with fishing df, and it worked
read_from_gsheets("https://docs.google.com/spreadsheets/d/1wXtfPvCVdUOg3r18jrpiOVmdM6WPMbaP72D6USnG74E/edit?usp=sharing").head()
```

```python
# importing the dataframes

df_fishing = read_from_gsheets("https://docs.google.com/spreadsheets/d/1wXtfPvCVdUOg3r18jrpiOVmdM6WPMbaP72D6USnG74E/edit?usp=sharing")
df_surfing = read_from_gsheets("https://docs.google.com/spreadsheets/d/1tA125bwnlMak3rQ4n5RfErFKrGg_eUJRjZ4H6LfIfWk/edit?usp=sharing")
```

```python
# now i want to merge my 'df_joined' with those, on 'month' and 'area'
# but to avoid ambiguity, i'll change the names of activity

# changing the columns names of fishing

new_fishing_columns = ['area','month','fishing_season']
df_fishing.columns = new_fishing_columns

# changing the columns names of surfing

new_surfing_columns = ['area','month','surfing_season']
df_surfing.columns = new_surfing_columns
```

```python
# merging done

df_joined_fishing = pd.merge(left=df_joined, right=df_fishing, how='left', on=['month', 'area'])
df_complete = pd.merge(left=df_joined_fishing, right=df_surfing, how='left', on=['month', 'area'])
df_complete.sample(10)
```

```python
# now i want to export some df's
# one with year, month, area, count(activity) fatal, fishing_season and surfing_season, for more exploratory analysis
# a grouped one, count of activity per year and month, filtered by each area

cols_wanted = ['year','month','area','activity','fatal_y_n','fishing_season','surfing_season']
df_export_exp = df_complete[cols_wanted]
df_export_exp.to_csv('complete_table.csv')
```

```python
# for the second export, there is a problem
# i want all years and months per area, even if there's no incident reported
# if i group by the year and month i cant get that result
# i'll try to create a dataframe with combinations of year, month and area, similar the one i did back at the start

# creating a new df from zero, with range of datetime

florida_df = pd.DataFrame()
florida_df['date'] = pd.date_range('01-01-2008','01-01-2018', freq='M')

# creating and year and month column in the test df

florida_df['year'] = florida_df['date'].dt.year
florida_df['month'] = florida_df['date'].dt.month

# adding the column area filled with 'florida'

florida_df['area'] = 'florida'
```

```python
# doing the same with california and hawaii

california_df = pd.DataFrame()
california_df['date'] = pd.date_range('01-01-2008','01-01-2018', freq='M')
california_df['year'] = california_df['date'].dt.year
california_df['month'] = california_df['date'].dt.month
california_df['area'] = 'california'

hawaii_df = pd.DataFrame()
hawaii_df['date'] = pd.date_range('01-01-2008','01-01-2018', freq='M')
hawaii_df['year'] = hawaii_df['date'].dt.year
hawaii_df['month'] = hawaii_df['date'].dt.month
hawaii_df['area'] = 'hawaii'
```

```python
# now i'll concatenate those 3, it worked

area_df = pd.concat([florida_df, california_df, hawaii_df])
```

```python
# merging with the df_complete

df_year_month = pd.merge(left=area_df, right=df_complete, how='left', on=['year', 'month', 'area'])

# testing the groupby, and it brought every month as desired

df_hawaii = df_year_month[df_year_month['area'] == 'hawaii'].groupby(['area','year','month'])[['activity']].count()
df_california = df_year_month[df_year_month['area'] == 'california'].groupby(['area','year','month'])[['activity']].count()
df_florida = df_year_month[df_year_month['area'] == 'florida'].groupby(['area','year','month'])[['activity']].count()
```

```python
# exporting df per area

df_hawaii.to_csv('incidents_hawaii.csv')
df_california.to_csv('incidents_california.csv')
df_florida.to_csv('incidents_florida.csv')
```
